```python
df_NVDI_Class = pd.read_csv("NVDIs of all cities - NVDI all cities.csv")
```

```python
df_NVDI_Class.head()
```

```python
mask = df_NVDI_Class['City'].isin(['Buenos Aires', 'Bogota', 'Sao Paulo', 'Mexico City'])
df_NVDI_Class[~mask]
```

```python
df_NVDI_Class.drop(df_NVDI_Class.index[255:419])
```

```python
df_filtered_NVDI_Class = df_NVDI_Class.drop(df_NVDI_Class.index[255:419])
```

```python
def categorize_nvdi(NVDI):
    if NVDI < 0.2:
        return "Nein"
    else:
        return "Ja"

# Apply the function to the Age column using the apply() function
df_NVDI_Class["NVDI above 0.2"] = df_NVDI_Class["NVDI"].apply(categorize_nvdi)
```

```python
df_NVDI_Class.head()
```

```python
df_NVDI_Class.to_csv("NVDI_Class.csv")
```

```python
def categorize_nvdi(NVDI):
    if NVDI < 0.2:
        return "Nein"
    else:
        return "Ja"

# Apply the function to the Age column using the apply() function
df_filtered_NVDI_Class["NVDI above 0.2"] = df_filtered_NVDI_Class["NVDI"].apply(categorize_nvdi)
df_filtered_NVDI_Class.head()
```
